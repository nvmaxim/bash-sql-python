```python
# Front
# УПРАЖНЕНИЕ 3: "Консолидация данных (python)
# Объедините данные за Q1 и Q2, сгруппируйте по менеджерам и посчитайте общую выручку для каждого менеджера за оба периода.
from _001_main import df_csv
import pandas as pd
import numpy as np

df = df_csv.copy()

table_q1 = df[df["period"] == "Q1"].reset_index(drop=True)
table_q2 = df[df["period"] == "Q2"].reset_index(drop=True)

combined = pd.concat([table_q1, table_q2])

result = (
    combined.groupby("manager")
    .agg(total_revenue=("revenue", "sum"))
    .sort_values("total_revenue", ascending=False)
    .reset_index()
)

print(result, sep="\n" * 2)
```
