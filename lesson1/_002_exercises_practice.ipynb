```python
# Front Как, не используя datetime и методы списков/строк, преобразовать целое число секунд seconds в формат дни:часы:минуты:секунды (например, 90061 → 1:1:1:1)?

sec = 90061

day = sec // 86400
sec %= 86400

hour = sec // 3600
sec %= 3600

minute = sec // 60
sec %= 60

print(day, hour, minute, sec, sep=":")
```
