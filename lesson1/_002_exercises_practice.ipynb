```python
a = "maabccdee"
b = "ddeffggaaz"
# Результат: "abcdefgz"

combined = a + b
unique = ""

for char in combined:
    if char not in unique:
        unique += char
sorted_result = ""
len_unique = 0
for _ in unique:
    len_unique += 1

while len_unique > 0:
    min_symbol = unique[0]
    for char in unique:
        if char < min_symbol:
            min_symbol = char
    sorted_result += min_symbol
    new_unique = ""
    removed = False
    for ch in unique:
        if not removed and ch == min_symbol:
            removed = True
            continue
        sorted_result += ch
    unique = new_unique

print(unique, len_unique, min_symbol, sep="\n")
```
